```python
df = df.sort_index()

train_percentage = 0.8
train_size = int(len(df) * train_percentage)

train_data = df[:train_size]
test_data = df[train_size:]
```

```python
color_pal=sns.color_palette()
plt.figure(figsize=(10, 5))
plt.plot(train_data['y'].index, train_data['y'].values, '.', ms=1, color=color_pal[0], label='Train Data')
plt.plot(test_data['y'].index, test_data['y'].values, '.', ms=1, color=color_pal[1], label='Test Data')
plt.title('Temperature')
plt.legend()
plt.show()
```

```python


model = Prophet()
model.add_regressor('rh2m')
model.add_regressor('ws10m')
model.add_regressor('ps')

df['ds'] = df['ds']
df['y'] = df['y']  
df['rh2m'] = df['rh2m']
df['ws10m'] = df['ws10m']
df['ps'] = df['ps']

df['y'] = df['y']+1
# params = {
#     'growth': 'linear',                     # Assuming a linear trend in temperature
#     'changepoints': None,                   # Let Prophet automatically select changepoints
#     'n_changepoints': 30,                  # Increase the number of potential changepoints
#     'changepoint_range': 0.95,              # Allow changepoints across most of the time series
#     'yearly_seasonality': True,             # Assuming yearly seasonal patterns
#     'weekly_seasonality': True,            # No significant weekly patterns (adjust as needed)
#     'daily_seasonality': True,              # Assuming daily seasonal patterns
#     'seasonality_mode': 'multiplicative',   # Seasonal effects may vary in magnitude
#     'seasonality_prior_scale': 5.0,         # Moderate flexibility for seasonal components
#     'holidays_prior_scale': 10.0,           # Default value for holidays (adjust if needed)
#     'changepoint_prior_scale': 0.1,         # Increase flexibility for trend changes
#     'mcmc_samples': 0,                   # Enable MCMC sampling for uncertainty estimation
#     'interval_width': 0.9,                  # Adjust the width of uncertainty intervals
#     'uncertainty_samples': 2000,            # Increase the number of uncertainty samples         # Give more weight to the regressors
#     'stan_backend': None                    # Use the default backend
# }
```

```python
fit_model = model.fit(train_data)
```

```python
forecast = model.predict(test_data)
```

```python
f, ax = plt.subplots(1)
f.set_figheight(5)
f.set_figwidth(15)
ax.scatter(test_data.index, test_data['y'], color='r')
fig = model.plot(forecast, ax=ax)
```

```python
fig_comp = model.plot_components(forecast)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score
import numpy as np

mae = mean_absolute_error(test_data['y'], forecast['yhat'])
mse = mean_squared_error(test_data['y'], forecast['yhat'])
rmse = np.sqrt(mse)
mape = np.mean(np.abs((np.array(test_data['y']) - np.array(forecast['yhat'])) / np.array(test_data['y']))) * 100
r2 = r2_score(test_data['y'], forecast['yhat'])
```

```python
print('mae :', mae)
print('mse :', mse)
print('rmse :', rmse)
print('mape :', mape)
print('r2score:', r2)
```
